```python
import seaborn as sns
import pandas as pd

dataframe = sns.load_dataset("diamonds")
dataframe
```

```python
print(dataframe.dtypes)
```

Cópia de segurança do Data frame antes de realizar as transformações:

<h2>Transformações</h2>
<h4>Transformando dólar para real:</h4>

`Atenção!` essa transformação obedece a cotação do dólar do dia 15/08/2023

```python
dataframe2 = dataframe.copy()
```

```python
dataframe2["price"] = dataframe2["price"] * 4.99
dataframe2
```

<h4>Transformando os comprimentos X, Y e Z, que estão em mm, para nm:</h4>

```python
dataframe2["x"] = dataframe2["x"] * 10**6
dataframe2["y"] = dataframe2["y"] * 10**6
dataframe2["z"] = dataframe2["z"] * 10**6
dataframe2
```

<h2>Normalizações</h2>
<h4>Normalização do atributo carat pelo método padrão:</h4>
A `normalização padrão`, que consiste em alterar a média dos dados para zero e o desvio padrão dos dados para um, obtemos o seguinte resultado:

```python
media = dataframe2['carat'].mean()
desvio_padrao = dataframe2['carat'].std()

dataframe2['Carat_zscore'] = (dataframe2['carat'] - media) / desvio_padrao

dataframe2
```

<h4>Normalização do atributo carat pelo método de máximo absoluto:</h4>
A <b>normalização pelo máximo absoluto</b> , que faz com que o valor máximo dos dados seja 1, obtemos o seguinte resultado:

```python
carat_maximo = max(abs(dataframe2['carat']))

dataframe2['Carat_norm_max_abs'] = dataframe2['carat'] / carat_maximo

dataframe2
```

<h4>Normalização do atributo carat pelo método de mínimo e máximo:</h4>
A <b>normalização pelo mínimo e máximo</b>, que modifica os dados de forma que, após a transformação, os dados tenham valor mínimo de 0 e valor máximo de 1, obtemos o seguinte resultado:

```python
maximo = dataframe2['carat'].max()
minimo = dataframe2['carat'].min()

dataframe2['Carat_norm_min_max'] = (dataframe2['carat'] - minimo) / (maximo - minimo)

dataframe2
```

<h2>Utilizando o módulo scikit-learn! :D </h2>
Realizando conversão simbólico-numérico de diferentes maneiras com o scikit-learn:<br>
<h4>One-hot encoder no atributo color:</h4>

```python
from sklearn.preprocessing import OneHotEncoder

x = dataframe2["color"].values.reshape(-1, 1)

encoder = OneHotEncoder(sparse=False)

encoder.fit(x)

colunas = encoder.categories_

dados_convertidos = encoder.transform(x)

idade_desnormalizada = encoder.inverse_transform(dados_convertidos)

for i, col in enumerate(colunas):
    dados = dados_convertidos[:,i]
    dataframe2[col] = dados

# print("Dados originais")
# print(x)
# print()

# print("Dados convertidos:")
# print(colunas)
# print(dados_convertidos)
# print()

# print("Dados desconvertidos:")
# print(idade_desnormalizada)
dataframe2
```

<h4>Ordinal encoder no atributo color:</h4>

```python
from sklearn.preprocessing import OrdinalEncoder

x = dataframe2["cut"].values.reshape(-1, 1)

ordem = [["Fair", "Good", "Very Good", "Premium", "Ideal"]]
encoder = OrdinalEncoder(categories=ordem)

encoder.fit(x)

dados_convertidos = encoder.transform(x)

idade_desnormalizada = encoder.inverse_transform(dados_convertidos)

# Add as modificações no dataframe
dataframe2["cut"] = dados_convertidos
dataframe2
```
